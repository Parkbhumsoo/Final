```
import pandas as pd
from plotnine import *
import seaborn as sns
import matplotlib.pyplot as plt
import folium
%matplotlib inline
```

```
df = pd.read_excel('./코인세탁소위경도.xlsx') #전처리 해놓은 파일 받기.
df
```

```
df1= df.drop(0,0)
df1
```

```
gu = pd.read_excel('./구별 인원.xlsx') #구별인원
gu
```

```
state_geo = 'https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json'  #서울시 구별 json
```

```
m = folium.Map(location=[37.562225, 126.978555], tiles="OpenStreetMap", zoom_start=11)   #지도 초기값 설정.
map = folium.Map(location=[37.562225, 126.978555], tiles="OpenStreetMap", zoom_start=11)
mep = folium.Map(location=[37.562225, 126.978555], tiles="Stamen Toner", zoom_start=11)
```

```
# for n in df1.index:
#     popup_name = df1.loc[n, 'name']  #이름 표기하려고했으나 한글표기 X
```

```
for maps in df1.index:   # 데이터의 위도경도를 이용한 시각화
  lat = df1.loc[maps,'Latitude']
  lng = df1.loc[maps,'Longitude']
  folium.Marker([lat,lng],
                icon=folium.Icon(color='green')).add_to(m)   #위치를 icon 으로 표시
```

```
m.choropleth(
    geo_data=state_geo,
    name='1인가구',
    data=gu,
    columns=['gu', 'gaguone'],  # json에 맞는 구 이어야함.(1열)
    key_on='feature.properties.name',
    fill_color='Blues',
    fill_opacity=0.7,
    line_opacity=0.3,
    color = 'gray',
    legend_name = '1_gagu'
)  # json 을 기반으로한 구역나눔 시각화
```

```
map.choropleth(
    geo_data=state_geo,
    name='2인가구',
    data=gu,
    columns=['gu', 'gagutwo'],
    key_on='feature.properties.name',
    fill_color='PuRd',
    fill_opacity=0.7,
    line_opacity=0.3,
    color = 'gray',
    legend_name = '2_gagu'
)
for maps in df1.index:
  lat = df1.loc[maps,'Latitude']
  lng = df1.loc[maps,'Longitude']
  folium.RegularPolygonMarker([lat,lng],
                fill_color='#45647d',
                number_of_sides=6,radius=5).add_to(map)  #위치를 각형으로 표시
```

```
from folium.plugins import MarkerCluster #클러스터해보기
mep.choropleth(
    geo_data=state_geo,
    name='2인가구',
    data=gu,
    columns=['gu', 'gagutwo'],
    key_on='feature.properties.name',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.3,
    color = 'gray',
    legend_name = '2_gagu'
)
marker_cluster = MarkerCluster().add_to(mep)  #클러스터링이 됌.
for maps in df1.index:
  lat = df1.loc[maps,'Latitude']
  lng = df1.loc[maps,'Longitude']
  folium.RegularPolygonMarker([lat,lng],
                fill_color='blue',
                number_of_sides=6,radius=5).add_to(marker_cluster)
```
