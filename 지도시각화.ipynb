```
import pandas as pd
from plotnine import *
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
import folium
```

```
df = pd.read_excel('./laundry3.xlsx')
df
```

```
df1= df.drop(0,0)
df1
```

```
gu = pd.read_excel('./구별 인원.xlsx')
gu
```

```
state_geo = 'https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json'
```

```
m = folium.Map(location=[37.562225, 126.978555], tiles="OpenStreetMap", zoom_start=11)
map = folium.Map(location=[37.562225, 126.978555], tiles="OpenStreetMap", zoom_start=11)
```

```
for n in df1.index:
    popup_name = df1.loc[n, 'name']
```

```
lng_list = []; lat_list = [];

for maps in df1.index:
  lat = df1.loc[maps,'Latitude']
  lng = df1.loc[maps,'Longitude']
  folium.Marker([lat,lng],
                popup=popup_name,
                icon=folium.Icon(icon='cloud')).add_to(m)
```

```
m.choropleth(
    geo_data=state_geo,
    name='1인가구',
    data=gu,
    columns=['gu', 'gaguone'],
    key_on='feature.properties.name',
    fill_color='Blues',
    fill_opacity=0.7,
    line_opacity=0.3,
    color = 'gray',
    legend_name = '1_gagu'
)
```

```
lng_list = []; lat_list = [];

for maps in df1.index:
  lat = df1.loc[maps,'Latitude']
  lng = df1.loc[maps,'Longitude']
  folium.Marker([lat,lng],
                popup=popup_name,
                icon=folium.Icon(icon='star')).add_to(map)

map.choropleth(
    geo_data=state_geo,
    name='2인가구',
    data=gu,
    columns=['gu', 'gagutwo'],
    key_on='feature.properties.name',
    fill_color='Reds',
    fill_opacity=0.7,
    line_opacity=0.3,
    color = 'gray',
    legend_name = '2_gagu'
)
```
