```python
from bs4 import BeautifulSoup as bs
from selenium import webdriver
import time
import numpy as np
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import os
import bs4
```

```python
os.getcwd()
```

```python
options = webdriver.ChromeOptions()  # 옵션 있어도되고 없어도 됌.
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
```

```python
driver=webdriver.Chrome('./chromedriver') #크롬드라이버를 쓴다.
driver.implicitly_wait(3)
driver.get('https://map.kakao.com/')  # 이 웹사이트로 들어간다.
driver.find_element_by_css_selector('#search\.keyword\.query').send_keys('셀프빨래방') #셀프빨래방을 입력
driver.find_element_by_xpath('//*[@id="search.keyword.submit"]').send_keys(Keys.ENTER) #엔터.
time.sleep(2) #2초 쉬기
a = driver.find_element_by_xpath('//*[@id="search.keyword.bounds"]') #아래1줄과 세트.
driver.execute_script("arguments[0].click()",a)   #클릭
clean_list = [];clean_list2=[]; clean_list3=[]; 
b = driver.find_element_by_xpath('//*[@id="info.search.place.more"]')
driver.execute_script("arguments[0].click()",b)
time.sleep(2)
c = driver.find_element_by_xpath('//*[@id="localInfo.map.county"]')
driver.execute_script("arguments[0].click()",c)
for i in range(1,26):  #구별로 보기 loop
    d = driver.find_element_by_xpath('//*[@id="localInfoList.county.list"]/li[%d]/a'%i)
    driver.execute_script("arguments[0].click()",d)
    for j in range(1,16):
        if j%5 == 0: #5페이지(끝페이지)이면 다음페이지로 누르기 loop
            c = driver.find_element_by_xpath('//*[@id="info.search.page.no%d"]'%j)
            driver.execute_script("arguments[0].click()",c)
            clean = bs4.BeautifulSoup(driver.page_source, 'lxml')
            cleaner = clean.find_all('a',class_='link_name')                #크롤링
            cleaner2 = clean.find_all('p',attrs = {'data-id': 'address'})    
            cleaner3 = clean.find_all('p',attrs = {'data-id': 'otherAddr'})  
            time.sleep(1)
            for clean in cleaner2:             #크롤링 후 리스트에 저장
                clean_list2.append(clean.text)
            for clean in cleaner:
                clean_list.append(clean.text)
            for clean in cleaner3:
                clean_list3.append(clean.text)    
            time.sleep(2)
            a = driver.find_element_by_xpath('//*[@id="info.search.page.next"]')
            driver.execute_script("arguments[0].click()",a)
        if j<5:
            b = driver.find_element_by_xpath('//*[@id="info.search.page.no%d"]'%j)
            driver.execute_script("arguments[0].click()",b)
            time.sleep(2)
            clean = bs4.BeautifulSoup(driver.page_source, 'lxml')
            cleaner = clean.find_all('a',class_='link_name')
            cleaner2 = clean.find_all('p',attrs = {'data-id': 'address'})
            cleaner3 = clean.find_all('p',attrs = {'data-id': 'otherAddr'})
            time.sleep(1)
            for clean in cleaner2:
                clean_list2.append(clean.text)
            for clean in cleaner:
                clean_list.append(clean.text)
            for clean in cleaner3:
                clean_list3.append(clean.text)
            time.sleep(2)
        else:
            break
```
