```python
from bs4 import BeautifulSoup as bs
from selenium import webdriver
import time
import numpy as np
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import os
import bs4
```

```python
os.getcwd()
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
```

```python
driver=webdriver.Chrome('./chromedriver')
driver.implicitly_wait(3)
driver.get('https://map.kakao.com/')
driver.find_element_by_css_selector('#search\.keyword\.query').send_keys('셀프빨래방')
driver.find_element_by_xpath('//*[@id="search.keyword.submit"]').send_keys(Keys.ENTER)
time.sleep(2)
clean_list = [];clean_list2=[];
a = driver.find_element_by_xpath('//*[@id="info.search.place.more"]')
driver.execute_script("arguments[0].click()",a)
time.sleep(2)
for i in range(1,34):
    for j in range(1,6):
        if j%5 == 0:
            c = driver.find_element_by_xpath('//*[@id="info.search.page.no%d"]'%j)
            driver.execute_script("arguments[0].click()",c)
            clean = bs4.BeautifulSoup(driver.page_source, 'lxml')
            cleaner = clean.find_all('a',class_='link_name')
            cleaner2 = clean.find_all('p',attrs = {'data-id': 'address'})
            for clean in cleaner2:
                clean_list2.append(clean.text)
            for clean in cleaner:
                clean_list.append(clean.text)
            time.sleep(2)
            a = driver.find_element_by_xpath('//*[@id="info.search.page.next"]')
            driver.execute_script("arguments[0].click()",a)
        else:
            b = driver.find_element_by_xpath('//*[@id="info.search.page.no%d"]'%j)
            driver.execute_script("arguments[0].click()",b)
            time.sleep(2)
            clean = bs4.BeautifulSoup(driver.page_source, 'lxml')
            cleaner = clean.find_all('a',class_='link_name')
            cleaner2 = clean.find_all('p',attrs = {'data-id': 'address'})
            for clean in cleaner2:
                clean_list2.append(clean.text)
            for clean in cleaner:
                clean_list.append(clean.text)
```
